```python
# importing liabraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# base url
base_url = "https://www.flipkart.com/search?q=best+mobile+under+20000rs&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D20000"
```

```python
# creating pagination till which page we want to scrape
urls=[base_url]
for i in range(2,61):
  urls.append(base_url+"&page="+str(i))
print(urls)
```

# extracting contents

```python
# requesting url
source = requests.get(base_url)
```

```python
# creating object
soup = BeautifulSoup(source.text,"html.parser")
```

```python
# all containers of page where product resides
container_page = soup.find_all("div", class_ = "_2kHMtA" )
```

```python
# finding name of mobile
container_page[0].find("div",class_ = "_4rR01T").text
```

```python
# finding mobile brand
container_page[0].find("div",class_ = "_4rR01T").text.split()[0]
```

```python
# finding mobile price
container_page[0].find("div", class_ ="_30jeq3 _1_WHN1").text.strip("₹")
```

```python
# finding mobile rating
container_page[0].find("div",class_ = "_3LWZlK").text
```

```python
# finding total rating
container_page[0].find("span", class_ = "_2_R_DZ").text.split()[0]
```

```python
# finding total reviews
container_page[0].find("span", class_ = "_2_R_DZ").text.split()[3]
```

```python
# scrapinng all contents and from all pages
data_list=[]
for url in urls:
  # requesting website to scrape the data
  source = requests.get(url)

  # creating a object soup
  soup = BeautifulSoup(source.text,"html.parser")

  # all containers of page where product resides
  container_page = soup.find_all("div", class_ = "_2kHMtA" )

  for i in container_page:
    micro_list=[]
    try:
      # mobile name
      micro_list.append(i.find("div",class_ = "_4rR01T").text)
      # mobile brand
      micro_list.append(i.find("div",class_ = "_4rR01T").text.split()[0])
      # mobile price
      micro_list.append(i.find("div", class_ ="_30jeq3 _1_WHN1").text.strip("₹"))
      # mobile rating
      micro_list.append(i.find("div",class_ = "_3LWZlK").text)
      # total ratings
      micro_list.append(i.find("span", class_ = "_2_R_DZ").text.split()[0])
      # total reviews
      micro_list.append(i.find("span", class_ = "_2_R_DZ").text.split()[3])

    except:
      micro_list.append(None)
    
    data_list.append(micro_list)

```

```python
print(data_list)
```

```python
print(len(data_list))
```

# creating dataframe from scraped data

```python
# creating dataframe
col = ["Mobile Name","Brand","Price","Rating","Total Rating","Total Review"]
df = pd.DataFrame(data_list,columns=col)
```
