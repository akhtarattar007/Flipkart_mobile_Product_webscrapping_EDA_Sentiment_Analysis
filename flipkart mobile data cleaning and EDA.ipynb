# EDA

### TOP 3 Brand who sold high quantity of mobiles

```python
df.columns
```

```python
df1 = df[["Brand","Price"]]
```

```python
df1.head()
```

```python
total_mobile_count_sold = df1.groupby(by = "Brand").count()
total_mobile_count_sold.head()
```

```python
# Top 3 brand who sold the most of mobiles
total_mobile_count_sold.sort_values(by="Price",ascending=False).head(3)
```

### Bottom 3 brands who sold less mobiles

```python
# Bottom 3 brand who sold the less  mobiles
total_mobile_count_sold.sort_values(by="Price",ascending=False).tail(3)
```

### Top 5 mobiles which highest rating counts

```python
df.columns
```

```python
df2 =df[['Mobile_Name','Total_Rating']].sort_values(by='Total_Rating',ascending =False)
df2.head()
```

# Bottom 5 mobile with less ratings counts

```python
# Bottom 5 mobiles with less reviews
df3 = df[["Mobile_Name","Total_Rating"]].sort_values(by="Total_Rating",ascending = True).head()
df3
```

### Top 3 Mobile brands  with their average rating

```python
df.columns
```

```python
df3 =  df[['Brand','Rating']].groupby(by='Brand').mean()
df3.sort_values(by="Rating",ascending = False).head(3)
```

### Bottom 3 Mobile brands  with their average rating

```python
df4 =  df[['Brand','Rating']].groupby(by='Brand').mean()
df4.sort_values(by="Rating",ascending = True).head(3)
```

### TOP 3 Brands who made highest revenue

```python
df.columns
```

```python
df5 = df[['Brand','Price']].groupby(by='Brand').sum()
df5.sort_values(by='Price',ascending = False).head(3)
```

### Bottom 3 Brands who made lowest revenue

```python
df6 = df[['Brand','Price']].groupby(by='Brand').sum()
df6.sort_values(by='Price',ascending = True).head(3)
```

### which product is sold most

```python
df7 = df[["Mobile_Name"]].value_counts()
df7.sort_values(ascending=False).head(1)
```

```python
most_sold_mobile = df7.sort_values(ascending=False).head(1).index[0][0]
```

```python
print(f'The most sold mobile is {most_sold_mobile}')
```
